# ***Buscando Iphone na Amazon***
Programa que busca por "Iphone" na loja da amazon, 
e cria uma planilha Excel com os dados.

```
#Program created by Pedro Henrique on Google Colab

#---------INSTALAÇÃO SIMULADOR NAVEGADOR----------------------------------------
!pip install openpyxl
!pip install furl
!pip install selenium
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin


#----------IMPORT DAS BIBLIOTECAS-----------------------------------------------
import bs4
from bs4 import BeautifulSoup
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from lxml.html.soupparser import fromstring
from openpyxl import Workbook
#-------------------------------------------------------------------------------

#----------url pesquisa "IPHONE"---------------------------------------------------------------------------
urlPag= "https://www.amazon.com.br/s?k=iphone&__mk_pt_BR=%C3%85M%C3%85%C5%BD%C3%95%C3%91&ref=nb_sb_noss_1"
#----------------------------------------------------------------------------------------------------------

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

browser = webdriver.Chrome('chromedriver',options=chrome_options)

browser.get(urlPag) #Get URL
codFont = BeautifulSoup(browser.page_source, "html.parser") #GET CÓDIGO FONTE DA PÁGINA


#após, avaliar elementos da página, e detectar padrão onde as informações estão escritas.
_padraoInfoTitulo = 'div[class="a-section a-spacing-medium"]' #PADRÃO TITULO
padraoInfoPreco = 'div[class="a-section a-spacing-medium"]' #PADRÃO PREÇO
tituloElementos = codFont.select(_padraoInfoTitulo) #buscando no código fonte as informações escritas
precoElementos = codFont.select(padraoInfoPreco) #buscando no código fonte as informações escritas


_titulo = 'div > h2 > a'
_price = 'div > div > a > span > span'

arquivo_excel = Workbook()
planilha = arquivo_excel.active
planilha.title = "IPHONE AMAZON"

planilha['A1'] = 'Produto'
planilha['B1'] = 'Valor'


cont = 0
cont2 = 2
items = len(tituloElementos)
while (cont < items):
  x = tituloElementos[cont]
  t = x.select(_titulo)[0]
  titulo = t.get_text()
  planilha.cell(row=cont2, column=1, value=titulo)
  try:
    z = precoElementos[cont]
    p = z.select(_price)[0]
    price = p.get_text()
  except:
    price = ''
  planilha.cell(row=cont2, column=2, value=price)
  cont = cont + 1
  cont2 = cont2 + 1


arquivo_excel.save("buscaIphoneAmazon.xlsx")
from google.colab import files
files.download('buscaIphoneAmazon.xlsx')
```
